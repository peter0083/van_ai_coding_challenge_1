```python
from PIL import Image
import os
```

```python
# loop through all png images in chenuen folder
# reference: https://stackoverflow.com/questions/10377998/how-can-i-iterate-over-files-in-a-given-directory

# convert chenuen first
target_folder = "tf_files/data/train/chenuen/"
directory = os.fsencode(target_folder)

index = 0
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith(".png"):
        index = index + 1
        print("converting png file", str(index))
        print(filename)
        im = Image.open(target_folder + str(filename))
        rgb_im = im.convert('RGB')
        print('created chenuen'+str(index)+'.jpg')
        rgb_im.save(target_folder + 'chenuen'+str(index+1)+'.jpg')

        continue
    else:
        continue
    print("conversion completed")
```

```python
# convert koei
target_folder2 = "tf_files/data/train/koei/"
directory2 = os.fsencode(target_folder2)

index2 = 0
for file2 in os.listdir(directory2):
    filename2 = os.fsdecode(file2)
    if filename2.endswith(".bmp"):
        index2 = index2 + 1
        print("converting bmp file", str(index2))
        print(filename2)
        im2 = Image.open(target_folder2 + str(filename2))
        rgb_im2 = im2.convert('RGB')
        print('created koei'+str(index2)+'.jpg')
        rgb_im2.save(target_folder2 + 'koei'+str(index2+1)+'.jpg')

        continue
    else:
        continue
    print("conversion completed")
```
